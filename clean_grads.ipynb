```python
import torch
```

```python
fn = "/u/scratch/s/schhina/grad_genes.txt"

with open(fn, 'r') as f:
    raw_dict = f.readlines()[0]
```

```python
print(raw_dict)
```

```python
entries = raw_dict.split(", device='cuda:0')")
```

```python
entries = filter(lambda x : x not in ",{} 'tensor(, device=')", raw_dict)
```

```python
semi_raw = ''.join(entries)
```

```python
entries = semi_raw.split('ua:0')
```

```python
entries[:-1]
```

```python
grads = {}
for entry in entries[:-1]:
    name, value = entry.split(":")
    if '-' in value:
        l, r = value.split('-')
        value = l + 'e-' + r
    grads[name] = value
```

```python
print(grads)
```

```python
grads = {k: float(v) for k, v in grads.items()}
```

```python
sorted_grads = sorted(((v, k) for k, v in grads.items()), reverse=True)
```

```python
sorted_grads[-5:] + sorted_grads[:5]
```

```python
names = [name for _, name in sorted_grads[-5:] + sorted_grads[:5]]
```

```python
s = f"{names}"
s = s[1:-1]
s = ''.join(filter(lambda x : x not in ",'", s))
s = ''.join(map(lambda c : c if c != ' ' else '\n', s))
print(s)
```
