```python
import pandas as pd

fn = "/u/scratch/s/schhina/AllPredictions.AvgHiC.ABC0.015.minus150.ForABCPaperV3.txt"
```

```python
df = pd.read_csv(fn, sep='\t', header=0)
```

```python
df.head()
```

```python
# True in this array means we want to keep the respective column
# In our case, we only want start, end and targetGene
cols = [ True, True, True, False, False, False, True, False, False,
       False, False, False, False, False, False, False, False, False,
       False, False, False, False, False, False]
```

```python
temp_df = df.loc[:, cols]
```

```python
temp_df.head()
```

```python
temp_df.shape
```

```python
# Add peak_id field which is a concatenation of chr, start and end
temp_df.loc[:, 'peak_id'] = temp_df['chr'] + '_' + temp_df['start'].astype(str) + '_' + temp_df['end'].astype(str)
```

```python
# Remove start, end and chr fields
temp_df = temp_df.drop(columns=['chr', 'start', 'end'])
```

```python
# Rename TargetGene to gene_short_name
temp_df.rename(columns={'TargetGene': 'gene_short_name'}, inplace=True)
```

```python
output_file = "/u/scratch/s/schhina/abc_peaks.csv"
temp_df.to_csv(output_file)
```
